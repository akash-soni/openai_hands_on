## Chat Completion API

```python
response = openai.Completion.create(model='text-curie-001',
                         prompt="what is the capital of India?",)
```

```python
response
```

```python
response = openai.Completion.create(model='text-curie-001',
                         prompt="Explain machine learning to a 5 year old?",
                         max_tokens=25,
                         )
```

```python
response["choices"]
```

```python
response = openai.Completion.create(model='text-curie-001',
                         prompt="top 4 famous cities in the world",
                         max_tokens=25,
                         temperature=0.6,           
                         )
response
```

```python
response = openai.Completion.create(model='text-curie-001',
                         prompt="top 4 famous cities in the world",
                         max_tokens=25,
                         temperature=0.6,    
                         n=5       
                         )
response
```

for the above example with N we have diffrent choices of output.

```python
response = openai.Completion.create(model='text-curie-001',
                         prompt="Act as an AI model and explain quantum computing",
                         max_tokens=40,
                         temperature=0.6,         
                         )
```

```python
print(response["choices"][0]["text"])
```
